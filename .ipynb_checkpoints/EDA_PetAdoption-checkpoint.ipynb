### The Adoption Age Frequency

```python
#plot the age frequency distribution of cats and dogs that were adopted
filtered_Dogs = petsPrepared_df[(petsPrepared_df['Adopted']==1) & (petsPrepared_df['Type'] == 1 )]
filtered_Cats = petsPrepared_df[(petsPrepared_df['Adopted']==1) & (petsPrepared_df['Type'] == 2 )]

def ageDistribution(df_Dog,df_Cat):
    fig, axs = plt.subplots(1, 2, figsize=(12, 4))
    axs[0].hist(df_Dog['Age'])
    axs[0].title.set_text('Dog Age Adoption Frequency')
    axs[0].set_xlabel('Age (months)')
    axs[1].hist(df_Cat['Age'],color = "red")
    axs[1].title.set_text('Cat Age Adoption Frequency')
    axs[1].set_xlabel('Age (months)')
    plt.show()
```

```python
ageDistribution(filtered_Dogs,filtered_Cats)

print('Measures of Central Tendancy','\n')
print('DOG')
print('Average Age Adopted:',filtered_Dogs['Age'].mean().round(0))
print('Standard Deviation:',np.round(filtered_Dogs['Age'].std(),0))
print('Mode:',filtered_Dogs['Age'].mode()[0])
print('\n')
print('CAT')
print('Average Age Adopted:',filtered_Cats['Age'].mean().round(0))
print('Standard Deviation:',np.round(filtered_Cats['Age'].std(),0))
print('Mode:',filtered_Cats['Age'].mode()[0])
```

A positively skewed distribution for both cat and dog adoption age where the mode is 2. The mode being the most frequently occuring value in the dataset, in this case, the age of 2 months for both cats and dogs. This indicates the younger the pet, the more likely its going to be adopted, and a smaller number of instances where people adopt older pets.

### Gender Adoption Rate

Which gender is being adopted more?

```python
labels = 'Male','Female'
male_Dogs = (filtered_Dogs['Gender']==1).sum()
female_Dogs = (filtered_Dogs['Gender']==2).sum()
dogs_list = [male_Dogs,female_Dogs]

male_Cats = (filtered_Cats['Gender']==1).sum()
female_Cats = (filtered_Cats['Gender']==2).sum()
cats_list = [male_Cats,female_Cats]

#pie chart of genders adopted
def genderPie(dogs:list,cats:list):
    fig, axs = plt.subplots(1, 2, figsize=(12, 4))
    axs[0].pie(dogs, labels=labels,autopct='%1.1f%%')
    axs[0].title.set_text('Dog Gender Adoption Rate')
    axs[1].pie(cats, labels=labels,autopct='%1.1f%%')
    axs[1].title.set_text('Cat Gender Adoption Rate')
    plt.show()
```

```python
genderPie(dogs_list,cats_list)
```

The pie chart shows the female gender gets adopted slightly more than the male for both cats and dogs.

### To understand if there is favouritism between breeds

Sometimes, potential adopters walk into a shelter with a specific breed in mind already. We can indentify this popularity of breeds with the appropriate visualization.

What are the number of pure breeds and mixed in the dataset?

```python
count_of_Breeds = petsPrepared_df.BreedPure.value_counts()
count_of_Breeds
```

We have 20% more mixed breeds than pure breeds in the dataset.

```python
#merge the first breed column breed1
breeds_df = petsPrepared_df.filter(['Type','Adopted','Breed1','Breed2','Color1','Color2','Color3'])
breeds_df = breeds_df.rename(columns={'Breed1':'BreedID'})
breeds_df = pd.merge(breeds_df,breedLabels_df,on=['Type','BreedID'],how='inner')
breeds_df = breeds_df.drop(columns={'BreedID'})
```

```python
breedCategories = breeds_df.groupby(['BreedName','Type']).count().sort_values('Adopted',ascending=False)[:20]
breedCategories = breedCategories.reset_index()
breedCategories.head(2)
```
