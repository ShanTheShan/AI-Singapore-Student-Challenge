## The Adoption Age Frequency

Data set has bins of [0,2] and 2. I'll merge them together

```python
petsPrepared_df['AgeBins'] = petsPrepared_df['AgeBins'].apply(lambda value:'[0,2]' if value == '2' else value)
```

```python
ageBinsGrouped = petsPrepared_df.groupby(['Adopted','AgeBins','Type']).count().reset_index().filter(['Adopted','AgeBins','Type','PetID'])
ageBinsGrouped = ageBinsGrouped.rename(columns={'PetID':'Count'}) 
```

```python
ageBinsGroupedByDog = ageBinsGrouped.query('Type == 1')
ageBinsGroupedByCat = ageBinsGrouped.query('Type == 2')
```

To sort the column by age bins, need to convert to integers.

```python
def sortBinsAsc(x):
    if x == '[0,2]':
        return 2
    elif x == '[3,6]':
        return 3
    elif x == '[6,12]':
        return 6
    elif x == '[12,24]':
        return 12
    elif x == '[24,60]':
        return 24
    elif x == '[60,255]':
        return 60
```

```python
ageBinsGroupedByDog['AgeBins'] = ageBinsGroupedByDog['AgeBins'].apply(lambda value: sortBinsAsc(value))
ageBinsGroupedByCat['AgeBins'] = ageBinsGroupedByCat['AgeBins'].apply(lambda value: sortBinsAsc(value))
```

```python
ageBinsGroupedByDog_Adopted = ageBinsGroupedByDog[(ageBinsGroupedByDog.Adopted == 1)].sort_values(by=['AgeBins'])
ageBinsGroupedByDog_UnAdopted = ageBinsGroupedByDog[(ageBinsGroupedByDog.Adopted == 0)].sort_values(by=['AgeBins'])

ageBinsGroupedByCat_Adopted = ageBinsGroupedByCat[(ageBinsGroupedByCat.Adopted == 1)].sort_values(by=['AgeBins'])
ageBinsGroupedByCat_UnAdopted = ageBinsGroupedByCat[(ageBinsGroupedByCat.Adopted == 0)].sort_values(by=['AgeBins'])
```

```python
ageDataDogsAdopted = ageBinsGroupedByDog_Adopted['Count'].to_list()

ageDataDogsUnAdopted = ageBinsGroupedByDog_UnAdopted['Count'].to_list()

ageDataCatsAdopted = ageBinsGroupedByCat_Adopted['Count'].to_list()

ageDataCatsUnAdopted = ageBinsGroupedByCat_UnAdopted['Count'].to_list()
```

```python
ageLabels = ['0-2','3-6','6-12','12-24','24-60','60-255']
```

```python
def ageDistribution(dog1,dog2,cat1,cat2,labels):
    fig, axs = plt.subplots(1, 2, figsize=(12, 4))
    axs[0].bar(labels,dog1)
    axs[0].bar(labels,dog2,bottom = dog1)
    axs[0].title.set_text('Dog Age Adoption Frequency')
    axs[0].set_xlabel('Age Bins')
    axs[0].legend(['Adopted','UnAdopted'])
    
    axs[1].bar(labels,cat1)
    axs[1].bar(labels,cat2,bottom=cat1)
    axs[1].title.set_text('Cat Age Adoption Frequency')
    axs[1].set_xlabel('Age Bins')
    axs[1].legend(['Adopted','UnAdopted'])
    
    plt.show()
```

# Dogs

```python
dogs_df = petsPrepared_df.filter(['Type','Adopted','Color1','Color2','Color3'])
dogs_df = dogs_df.query('Type == 1 & Adopted == 1')
dogs_df.head()
```

```python
#function to get the color of a column in a df
def getColors(df,col):
    temp_df = df.rename(columns={col:'ColorID'})
    temp_df = pd.merge(temp_df,colorLabels_df,on=['ColorID'],how='left')
    temp_df = temp_df.drop(columns={'ColorID'})
    return temp_df
```

```python
#merge the color label csv 3 times
dogColors_df = getColors(dogs_df,'Color1')
dogColors_df = getColors(dogColors_df,'Color2')
dogColors_df = getColors(dogColors_df,'Color3')
```

```python
#split into 3 df, singular color, doubles and triples
colorList = colorLabels_df['ColorName'].tolist()

singleDogColor_df = dogColors_df[(dogColors_df['ColorName_x'].isin(colorList))
                                   &(dogColors_df['ColorName_y'].isna())
                                   &(dogColors_df['ColorName'].isna())]

doubleDogColor_df = dogColors_df[(dogColors_df['ColorName_x'].isin(colorList))
                                   &(dogColors_df['ColorName_y'].isin(colorList))
                                   &(dogColors_df['ColorName'].isna())]

tripleDogColor_df = dogColors_df[(dogColors_df['ColorName_x'].isin(colorList))
                                   &(dogColors_df['ColorName_y'].isin(colorList))
                                   &(dogColors_df['ColorName'].isin(colorList))]
```

```python
#groupyby function
def groupByColorColumns(df,threshold,*columns):
    cols = []
    cols.extend([i for i in columns])
    
    #for triple colors
    if len(cols) == 4:
        new_df = df.groupby(cols).count()
        #filter only dogs colors that were adopted more than 20 times, to remove low color combination counts
        new_df = new_df[new_df['Type']>threshold]
        return new_df
    
    #for double colors
    elif len(cols) == 3:
        new_df = df.drop(columns={'ColorName'}).groupby(cols).count()
        new_df = new_df[new_df['Type']>threshold]
        return new_df
    
    #for single colors
    elif len(cols) == 2:
        new_df = df.drop(columns={'ColorName_y','ColorName'}).groupby(cols).count()
        new_df = new_df[new_df['Type']>threshold]
        return new_df
```

### Adoption Rate for Dogs with Single Colored Coats

```python
singleDogColor_df = groupByColorColumns(singleDogColor_df,20,'Adopted','ColorName_x')
singleDogColorY = singleDogColor_df['Type'].tolist()
singleDogColorX = ['Black','Brown','Cream','Golden','Gray','White','Yellow']
```

```python
def colorPie(data, labels, colors, title,legend):
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(5, 5), gridspec_kw={'width_ratios': [4, 1]})
    wedges, texts = ax1.pie(data, labels=labels, colors=colors,textprops={'color': 'white'})
    ax1.set_title(title,fontsize=15,color='white')
    #background color
    fig.patch.set_facecolor('#2E2E2E')

    #hide the second subplot's axes (where the legend will be)
    ax2.axis('off')
    ax2.legend(wedges, legend, loc="center",labelcolor='white',facecolor='green')

    plt.tight_layout()
    plt.show()
```

The table above shows, we have 5 major color combinations for dogs being adopted with double colored coats, black, brown, cream, golden and gray.

```python
#function to extract the 5 main color data counts
def getDoubleCounts(color,df):
    x = df.loc[(1,color)]
    x = x.reset_index()
    x['ColorName_y'] = x['ColorName_y'].apply(lambda value: color+' '+value)
    #return a list of colors combinations and each combination counts
    colorCombi = x['ColorName_y'].tolist()
    counts = x['Type'].tolist()
    
    #convert labels with %
    total = sum(counts)
    for i in range(len(counts)):
        temp = counts[i]/total * 100
        colorCombi[i] = colorCombi[i] + '\n' + str(round(temp, 1)) + '%'
    
    return colorCombi,counts
```

```python
blackCombi_Dog,blackCombi_Dog_Counts = getDoubleCounts('Black',doubleDogColor_df)
brownCombi_Dog,brownCombi_Dog_Counts = getDoubleCounts('Brown',doubleDogColor_df)
```

```python
import squarify

def plotTreeMap(shape,title,size,labels):
    plt.figure(figsize=shape)
    plt.title(title,pad=20)
    plt.axis('off')
    squarify.plot(sizes=size,label=labels,pad=0, color=sns.color_palette("coolwarm",len(size)),alpha=0.6)
```

```python
#function to extract the 5 main color data counts
def getTripleCounts(color,df):
    x = df.loc[(1,color)]
    x = x.reset_index()
    x['ColorNames'] = color + x['ColorName_y'] + x['ColorName']
    x = x.drop(columns={'ColorName_y','ColorName'})
   
    x['ColorName'] = x['ColorNames'].apply(lambda value: re.sub('(?<=[a-z])(?=[A-Z])',',',str(value)))
    x = x.drop(columns={'ColorNames'})
    
    #return a list of colors combinations and each combination counts
    colorCombi = x['ColorName'].tolist()
    counts = x['Type'].tolist()
    
    #convert labels with %
    total = sum(counts)
    for i in range(len(counts)):
        temp = counts[i]/total * 100
        colorCombi[i] = colorCombi[i] + '\n' + str(round(temp, 1)) + '%'
    
    return colorCombi,counts
```

# Size of animals adopted

```python
def filterSizes(df,animal):
    temp = df[(df['Type'] == animal) & 
              (df['MaturitySize'].notnull()) & 
              (df['Adopted'] == 1)]
    
    temp = temp.filter(['Type','MaturitySize','Adopted'])
    #groupby size and count
    temp_grouped = temp.groupby(['MaturitySize','Adopted']).count()
    temp_grouped = temp_grouped.reset_index()
    
    return temp_grouped
```

```python
dogSizes_df = filterSizes(petsPrepared_df,1)
```

```python
def convertLabelToString(labels,type_):
    if type_ == 'size':
        SizeNames = ['Small','Medium','Large','Extra Large']
        for i in range(4):
            labels[i] = SizeNames[i]
        return labels
    
    elif type_ == 'fur':
        furTypes = ['Short','Medium','Long']
        for i in range(3):
            labels[i] = furTypes[i] 
        return labels
```

```python
dogSizesX = dogSizes_df['MaturitySize'].tolist()
dogSizesY = dogSizes_df['Type'].tolist()
```

```python
dogSizesX = convertLabelToString(dogSizesX,'size')
```

```python
catSizes_df = filterSizes(petsPrepared_df,2)
```

```python
catSizesX = catSizes_df['MaturitySize'].tolist()
catSizesY = catSizes_df['Type'].tolist()
```

```python
catSizesX = convertLabelToString(catSizesX,'size')
```

```python
def plotBarDouble(x1,y1,x2,y2,title1,title2):
    bar_colors = ['tab:orange', 'tab:green', 'tab:blue', 'tab:red']

    fig, ax = plt.subplots(1,2,figsize=(11,5))

    ax[0].bar(x1,y1,color=bar_colors)
    ax[0].set_title(title1)
    ax[0].set_xticks(np.arange(len(x1)))

    ax[1].bar(x2,y2,color=bar_colors)
    ax[1].set_title(title2)
    ax[1].set_xticks(np.arange(len(x2)))
    plt.show()
```

```python
plotBarDouble(dogSizesX,dogSizesY,catSizesX,catSizesY,'Size of Dogs being Adopted','Size of Cats being Adopted')
```

The bar charts above show that, ***medium sized dogs*** are heavily preferred by adopters looking for a dog, while both ***small and medium sized cats*** are being chosen by adopters. Large and extra large cats and dogs are not a popular choice amongst adopters.

# Adoption Figures for Sick/Injured Animals

```python
def filterHealth(df,animal,adopted):
    if adopted == 'yes':
        temp = df[(df['Type'] == animal) & 
                  (df['Health'].notnull()) & 
                  (df['Adopted'] == 1)]

        temp = temp.filter(['Type','HealthName','Adopted'])
        #groupby size and count
        temp_grouped = temp.groupby(['HealthName','Adopted']).count()
        temp_grouped = temp_grouped.reset_index()

        return temp_grouped
    
    else:
        temp = df[(df['Type'] == animal) & 
                  (df['Health'].notnull()) & 
                  (df['Adopted'] == 0)]

        temp = temp.filter(['Type','HealthName','Adopted'])
        #groupby size and count
        temp_grouped = temp.groupby(['HealthName','Adopted']).count()
        temp_grouped = temp_grouped.reset_index()

        return temp_grouped
```

```python
def filterVaccination(df,animal,adopted):
    if adopted == 'yes':
        temp = df[(df['Type'] == animal) & 
                  (df['Health'] == 1) & 
                  (df['Adopted'] == 1)]
        
        temp = temp.filter(['Adopted','VaccinatedName','DewormedName','SterilizedName'])
        temp_grouped = temp.groupby(['Adopted','VaccinatedName','DewormedName','SterilizedName']).size().reset_index()
        #filter by more than 100 dogs being adopted from the 3 combinations of health statuses
        temp_grouped = temp_grouped[temp_grouped[0]>100]
        
        return temp_grouped.sort_values(0,ascending=False)
    else:
        temp = df[(df['Type'] == animal) & 
                  (df['Health'] == 1) & 
                  (df['Adopted'] == 0)]
        
        temp = temp.filter(['Adopted','VaccinatedName','DewormedName','SterilizedName'])
        temp_grouped = temp.groupby(['Adopted','VaccinatedName','DewormedName','SterilizedName']).size().reset_index()
        #filter by more than 100 dogs being adopted from the 3 combinations of health statuses
        temp_grouped = temp_grouped[temp_grouped[0]>100]
        
        return temp_grouped.sort_values(0,ascending=False)
```

## Listing Description

```python
def ascii_lower(str_):
    #this regex removes characters that come after an apostrophe,eg apple's -> apple
    reg = re.sub(r"'[a-z]+",'',str_)
    x = reg.encode("ascii", "ignore").decode()
    x = x.lower()
    return x

def removeDigits(str_):
    #replace all special characters, that is, everything that isnt a letter or digit
    reg = re.sub(r'[^A-Za-z ]+','',str_)
    return reg
```

```python
petsPrepared_df['Description'] = petsPrepared_df['Description'].apply(ascii_lower)
petsPrepared_df['Description'] = petsPrepared_df['Description'].apply(removeDigits)
```

```python
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer, PorterStemmer
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk import FreqDist
from wordcloud import WordCloud
```

```python
tokenizedData= []

#define lemmatization object
lt = WordNetLemmatizer()

#create our stopwords set
myStopWords = set(stopwords.words('english'))

def custom_tokenizer(row):
    tokens = []
    if row['Adopted'] == 1:
        for sent in sent_tokenize(row['Description']):
            for word in word_tokenize(sent):
                if word not in myStopWords:
                    tokenizedData.append(word)
                    word = lt.lemmatize(word)
                    tokens.append(word)
            
        newDesc = ' '.join(tokens)
        return newDesc
    else:
        return row['Description']
```

```python
petsPrepared_df['clean_Description'] = petsPrepared_df.apply(custom_tokenizer,axis=1)
```

# Summary of Key Insights Discovered

To summarize, below are the insights that I have discovered from exploring this dataset, in helping to answer the problem statement of predicting adoption rate and understanding adopter's preference.

## Average Adoption Speed

The average adoption speed for animals at PetFinders.my is 2.5 months, with a standard deviation of 1 month. This indicates that the adoption speeds for most animals typically vary by no more than 1 month from the average.


## Age

Dogs are primarily being adopted between the age ranges of 0-2 months, with a small portion of adopter's adopting dogs between 2-5 years old, suggesting that a small percentage of adopters are willing to adopt older dogs.

Cats however, are primarily being adopted between 0-6 months. Beyond that, adoption rate for older cat tapers off significantly. Kittens are targeted the most.


## Gender

Adopters prefer female pets slightly more than male.

## Breeds

Mixed breed dogs are adopted the most, whilst domestic short and medium hair for cats. Natural for a pet welfare organization as the pure breed counts are low.

## Color

Brown dogs are being adopted the most, whilst the most popular color combination amongst dog adopters are black-brown, brown-white, black-brown-white.

White, black, and grey cats are the most popular choices among adopters. Those who prefer multicolored cats tend to choose combinations like black and white, black-grey-white, and black-brown-white.

## Size

Mediums sized dogs are adopted the most, while small and medium sized cats are adopted the most. Large dogs and cats are not favoured.

## Fur length

Pets with short furs are being adopted the most, while those with long furs are being adopted the least.

## Health status

Adopters are open to adopting both healthy and unhealthy animals. When grouped by health status, the proportion of adopted animals always surpasses that of unadopted ones, indicating a willingness to provide homes regardless of the animal's health condition.

For dogs, vaccination, deworming, and sterilization don't seem to be major concerns for adopters. The highest adoption counts are seen in both dogs that have received these treatments and those that haven't.

For cats, both adopted and unadopted animals are generally not vaccinated, dewormed, or sterilized. This aligns with the age of the cats being adopted, suggesting that most are kittens.

## Number of photos and vidoes in a pet's listing

Videos do not contribute to aiding in a pet's adopted speed at all. Dogs only require 1-3 photos, while cats require 1-5, suggesting adopters wish to see more photos of cats over dogs.
